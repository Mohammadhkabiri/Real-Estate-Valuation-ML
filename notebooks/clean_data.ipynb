```python
import pandas as pd
%load_ext autoreload
%autoreload 2
import sys
import os
sys.path.append(os.path.abspath('..'))
from src.persian_to_english import persian_to_english_digits
raw_home_data = pd.read_csv("../divar_full_data.csv")
raw_home_data.head()
raw_home_data.info()
raw_home_data.describe(include='all')
raw_home_data.head(20)
```

```python
cols_to_convert = ["Area", "Year", "Rooms", "Floor"]
for col in cols_to_convert:
    raw_home_data[col] = raw_home_data[col].apply(persian_to_english_digits)
    
raw_home_data["Area"] = pd.to_numeric(raw_home_data["Area"], errors="coerce")
raw_home_data["Year"] = pd.to_numeric(raw_home_data["Year"], errors="coerce")
raw_home_data["Rooms"] = pd.to_numeric(raw_home_data["Rooms"], errors="coerce")



for col in ["Parking", "Storage", "Elevator"]:
    raw_home_data[col] = raw_home_data[col].fillna(False).astype(bool)
    
raw_home_data.head()

raw_home_data.info()
raw_home_data["Floor"].value_counts(dropna=False).head(20)
```

```python
raw_home_data.describe(include="all")
```

```python
import re
import numpy as np
def extract_floor_number(text):
    if pd.isna(text):
        return np.nan
    text = str(text).strip()
    if 'همکف' in text and 'زیر' not in text:
        return 0.0
    if 'زیر' in text: 
        return -1.0
        
    match = re.search(r'\d+', text)
    if match:
        return float(match.group())
        
    return np.nan
raw_home_data['Floor'] = raw_home_data['Floor'].apply(extract_floor_number)
```

```python
raw_home_data.head(20)
```

```python
clean_data = raw_home_data.dropna(subset=['Area', 'Year', 'Rooms', 'Floor']).copy()
clean_data.reset_index(drop=True, inplace=True)
clean_data.info()
clean_data.head(20)
```

```python
current_year = 1405
clean_data['Age'] = current_year - clean_data['Year']
clean_data['Price_per_meter'] = clean_data['Exact_Price'] / clean_data['Area']
clean_data = clean_data.drop(columns=['Year'])
clean_data.info()
clean_data.head(20)
```

```python
import pandas as pd
import numpy as np

def remove_outliers_by_district(group):
    Q1_price = group['Price_per_meter'].quantile(0.25)
    Q3_price = group['Price_per_meter'].quantile(0.75)
    IQR_price = Q3_price - Q1_price
    lower_price = Q1_price - 1.5 * IQR_price
    upper_price = Q3_price + 1.5 * IQR_price

    Q1_area = group['Area'].quantile(0.25)
    Q3_area = group['Area'].quantile(0.75)
    IQR_area = Q3_area - Q1_area
    lower_area = Q1_area - 1.5 * IQR_area
    upper_area = Q3_area + 1.5 * IQR_area

    filtered_group = group[
        (group['Price_per_meter'] >= lower_price) &
        (group['Price_per_meter'] <= upper_price) &
        (group['Area'] >= lower_area) &
        (group['Area'] <= upper_area) &
        (group['Exact_Price'] >= 500_000_000) # حذف قیمت‌های به شدت پایین (زیر 500 میلیون)
    ]
    return filtered_group

final_data = clean_data.groupby('District').apply(remove_outliers_by_district)
final_data = final_data.reset_index(drop=True)
final_data.to_csv('cleaned_divar_data_grouped.csv', index=False, encoding='utf-8-sig')
print(f"All Data: {len(final_data)}")
```
